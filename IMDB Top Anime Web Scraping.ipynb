```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url="https://www.imdb.com/list/ls057577566/"
r=requests.get(url)
soup=BeautifulSoup(r.content,'html.parser')
```

```python
anime_name=[]
year=[]
rating=[]
genre=[]
description=[]


```

```python
anime_data=soup.findAll('div',class_='lister-item mode-detail')
```

```python
for store in anime_data:
    name=store.h3.a.text
    anime_name.append(name)
    year_of_release=store.h3.find('span',class_="lister-item-year text-muted unbold").text.replace('(','').replace(')','')
    year.append(year_of_release)
    ratings=store.find('span',class_='ipl-rating-star__rating').text
    rating.append(ratings) 
    genres=store.find('span',class_='genre').text
    genre.append(genres)
    descript=store.find('div',class_='list-description').text
    description.append(descript)
    
```

```python
anime_df=pd.DataFrame({'Anime Name':anime_name,'Year':year,'Genre':genre,'Rating':rating,'Description':description}
)
```
